```python
import pandas as pd
import urllib.request
import zipfile
import video_utils as vu
import media_pipe_handler as mph

download_url = "https://download.microsoft.com/download/3/c/a/3ca92c78-1c4a-4a91-a7ee-6980c1d242ec/MS-ASL.zip"
dataset_path = "msasl_dataset"
destination_path = "msasl.zip"

urllib.request.urlretrieve(download_url, destination_path)

with zipfile.ZipFile(destination_path, 'r') as zip_ref:
    zip_ref.extractall("msasl_dataset")

MSASL_train = pd.read_json("{}/MS-ASL/MSASL_train.json".format(dataset_path))
```

```python
for i in range(len(MSASL_train)):
    video = MSASL_train.iloc[i]
    video_path = vu.download_clip(video)
    if video_path is not None:
        print("Downloaded video {}".format(video_path))
        mph.extract_hand_keypoints(video_path=f"clips/{video_path}.mkv", output_dir=f"output_keypoints")
        vu.delete_clip(video_path)
```
